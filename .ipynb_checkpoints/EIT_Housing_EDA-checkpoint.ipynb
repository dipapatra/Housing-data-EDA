## Data Dictionary:

MLS: MLS stands for Multiple Listing Service, a database used by real estate agents to list properties for sale.

Sold_Price: This column represents the price at which price the property was sold.

Zipcode: This column captures the postal code or ZIP code of the property's location.

Longitude and Latitude: These columns provide the precise geographical coordinates of the property's location, usually in degrees.

Lot_Acres: This column indicates the size of the property's lot in acres. 

Taxes: It represents the annual tax payment required for the property.

Year_Built: This column represents the year in which the property was constructed or built.

Bedrooms: It represents no. of bedrooms the property has.

Bathrooms:  It represents no. of bathrooms the property has.

Square_Foot: It represents ttotal area of the property in square feet.

Garage: This column indicates no. of garege the property has.

Kitchen_Features: This field indicates the various features the property has.

Fireplaces: This column indicates no. of fireplaces the property have.

Floor_Type: This indicates the typesof flooring the property have. 

HOA: HOA stands for Homeowners Association. This column captures information about any fees or dues associated with the property's membership in a homeowners association.

sold_price and bathrooms show a positive correlation of 0.327, indicating a moderate positive association between the number of bathrooms in a house and its sold price.

sold_price and fireplaces have a positive correlation of 0.385. This suggests a moderate positive relationship between the presence of fireplaces in houses and their sold prices.

sold_price and bedrooms exhibit a positive correlation of 0.115. This indicates a weak positive relationship between the number of bedrooms in a house and its sold price.

sold_price and garage show a positive correlation of 0.094. This suggests a weak positive relationship between the number of garage spaces in a house and its sold price.

sold_price and latitude demonstrate a positive correlation of 0.426. This indicates a moderate positive association between the latitude of the house's location and its sold price.

sold_price and HOA have a positive correlation of 0.160. This suggests a weak positive relationship between the presence of HOA and the sold price of houses.

#### Box Plot

```python
# box plot analysis for below columns for box plots
columns = ['sold_price', 'taxes', 'bathrooms', 'sqrt_ft','toal_kit_features']

# Create box plots for each column
for col in columns:
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot: {col}')
    plt.xlabel(col)
    plt.show()
    
    # Calculate the first quartile (Q1)
    Q1 = df[col].quantile(0.25)

    # Calculate the third quartile (Q3)
    Q3 = df[col].quantile(0.75)

    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1

    # Calculate the lower bound and upper bound for outliers
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Identify outliers using the bounds
    outliers = df[(df['sold_price'] < lower_bound) | (df['sold_price'] > upper_bound)]

    # Print the outliers
    print(outliers)
  
```

```python
# Century wise counts

# Group the data by century and count the number of occurrences in each group
century_counts = df.groupby((df['year_built']//100)*100)['year_built'].count()

# Plot the counts using a bar plot
century_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Century')
plt.ylabel('Count')
plt.title('Number of Houses Built in Each Century')
plt.show()
```

```python
# Calculating the Average sold price century wise
# Group the data by century and calculate the average sold price
century_avg_price = df.groupby((df['year_built']//100)*100)['sold_price'].mean()

# Print the result
print(century_avg_price)
```

```python
#Ploting the average sold price
century_avg_price.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Century')
plt.ylabel('Average Sold Price')
plt.title('Average Sold Price by Century')
formatter = mticker.StrMethodFormatter('${x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```
